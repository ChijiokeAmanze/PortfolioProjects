```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

DATA IMPORT

```python
pet_df = pd.read_csv("pet_supplies_2212.csv")
pet_df.head()
```

## Task 2

DATA CLEANING

```python
pet_df.isnull().sum()
```

```python
pet_df.groupby("category")["rating"].median()
```

```python

for index, row in pet_df.iterrows():
    category = row["category"]
    if category == "Accessory":
        pet_df.at[index, "rating"] = 4.0
    elif category == "Equipment":
        pet_df.at[index, "rating"] = 5.0
    elif category == "Food":
        pet_df.at[index, "rating"] = 5.0
    elif category == "Housing":
        pet_df.at[index, "rating"] = 4.0
    elif category == "Medicine":
        pet_df.at[index, "rating"] = 6.0
    elif category == "Toys":
        pet_df.at[index, "rating"] = 6.0
    else:
        pet_df.at[index, "rating"] = 5
        

```

```python
pet_df['category'].value_counts()
```

```python
pet_df['size'] = pet_df['size'].str.replace("Small|SMALL", "small")
pet_df['size'] = pet_df['size'].str.replace("Medium|MEDIUM", "medium")
pet_df['size'] = pet_df['size'].str.replace("LARGE|Large", "large")
pet_df['size'].value_counts()
```

```python
pet_df['category'].value_counts()
```

```python
pet_df["item"] = pet_df["category"].apply(lambda x: "luxury" if x == "Toys" else "other")
```

## Task 3

DATA EXPLORATION

```python
sns.countplot(x = "size", data = pet_df)
```

The count plot for the "size" column shows that the majority of products are labeled as "small," followed by "medium" and "large." This suggests that smaller-sized products are more prevalent in the dataset

```python
sns.boxplot(x = "item", y = "sales", data = pet_df)
```

The box plot comparing "item" (luxury vs. other) with "sales" reveals interesting insights. Luxury items (categorized as "Toys") tend to have higher sales compared to other items. Additionally, the box plot shows that there are some outliers with exceptionally high sales in the "Toys" category, indicating potential opportunities for further analysis to identify what drives these high sales figures.

## Task 4

```python
sns.barplot(x = "category", y = "repeat_purchase", data = pet_df)
```

The bar plot illustrates the relationship between "category" and "repeat_purchase." It shows that the "Toys" category has the highest proportion of repeat purchases among all categories. This insight implies that customers are more likely to repeatedly purchase items in the "Toys" category, indicating a potential focus area for increasing sales through repeat customers

```python
sns.histplot(x = "sales", data = pet_df)
```

The histogram depicting the distribution of sales provides a sense of the overall sales distribution. It appears that most products have relatively low to moderate sales, with a peak around a certain sales range. However, there are a few products with exceptionally high sales values, represented as outliers on the right side of the distribution. These outliers may be worth investigating further to understand the factors contributing to their high sales.

```python
sns.barplot(x = "repeat_purchase", y = "sales", data = pet_df)
```

The bar plot displaying the relationship between "repeat_purchase" and "sales" indicates that products with repeat purchases tend to have higher sales on average compared to products without repeat purchases. This suggests that encouraging repeat purchases could be a strategy for boosting overall sales.
